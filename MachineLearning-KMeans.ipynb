```python
from scipy import misc
f = misc.face()

import matplotlib.pyplot as plt
plt.imshow(f)
plt.show()
```

```python
print(f.shape)
#rescale the colors so they lie between 0 and 1
f_reshape = (f / 255.0).reshape(-1,3) 
print(f_reshape.shape)
```

```python
from sklearn.cluster import KMeans
k_colors = KMeans(n_clusters=5).fit(f_reshape) 
y_pred=k_colors.predict(f_reshape) 
```

```python
print(k_colors.cluster_centers_.shape)
print(k_colors.labels_.shape)
newimg=k_colors.cluster_centers_[k_colors.labels_]
print(newimg.shape)

import numpy as np
newimg=np.reshape(newimg, (f.shape))
print(newimg.shape)
```

```python
#Now let's plot the two images side and by side and examine the differences. 
%matplotlib inline
fig = plt.figure(figsize=(10,10)) 
ax=fig.add_subplot(1,2,1,xticks=[],yticks=[],title='Original Image') 
ax.imshow(f) 
ax=fig.add_subplot(1,2,2,xticks=[],yticks=[],title='Color Compressed Image using K-Means') 
ax.imshow(newimg) 
plt.show() 
```
